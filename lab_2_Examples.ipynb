```python
# import packages
import statistics 
import pandas as pd
import scipy.stats as stats
from scipy.stats import sem
from statistics import mean, stdev
import matplotlib.pyplot as plt
import numpy as np
#include if using a jupyter notebook, remove if using a .py file
%matplotlib inline 
```

#### 2B:

```python
# data
Hot = [5.5, 4.2]
Room_temp = [5, 3.9]
Cold = [4, 3]
```

```python
# find the mean using the mean() function from the statistics library
Hot_mean = mean(Hot)
Room_temp_mean = mean(Room_temp)
Cold_mean = mean(Cold)


# find the standard deviation using the stdev() function from the statistics library
Hot_stdev = stdev(Hot)
Room_temp_stdev = stdev(Room_temp)
Cold_stdev = stdev(Cold)

# find the standard error using SciPY
Hot_sem = sem(Hot)
Room_temp_sem = sem(Room_temp)
Cold_sem = sem(Cold)
```

```python
# build a bar plot
plt.bar(['Hot (>42C)', 'Room temp (25C)', 'Cold (4C)'],[Hot_mean, Room_temp_mean, Cold_mean], yerr=[Hot_sem, Room_temp_sem, Cold_sem], capsize=10)
plt.xlabel('Temperature (degrees C)')
plt.ylabel('Distance Measured (mm)')
plt.title('2B: Effect of Temperature on Diffusion of KMnO4')
plt.show()
```
